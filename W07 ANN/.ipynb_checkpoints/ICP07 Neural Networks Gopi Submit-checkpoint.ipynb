## MNIST machine learning exercise

In this exercise we will compare the performance of three different modeling approaches at predicting handwritten numbers. 

We use the MNIST data set;

## Load libraries

```python
from __future__ import print_function
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import clear_output


np.random.seed(1)
```

## Load data and explore/get to know the data structure



Load the MNIST digits dataset. It's originally from UCI machine learning library, but included in SKLearn.

```python
mnist = datasets.load_digits() # sklearn includes this data set .. https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits
```

## Split data into training and test sets


```python
X_train, X_test, y_train, y_test = train_test_split(mnist.data, mnist.target, test_size=0.2, random_state=1)
```

### Neural Net with default parameters

```python
%%time

ann = MLPClassifier(hidden_layer_sizes=(60,50,40), solver='adam', max_iter=200)
_ = ann.fit(X_train, y_train)
```

```python
%%time
y_pred = ann.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

## NN With RandomizedSearchCV

```python
%%time

score_measure = "accuracy"
kfolds = 5

param_grid = {
    'hidden_layer_sizes': [ (50,), (70,),(50,30), (40,20), (60,40, 20), (70,50,40)],
    'activation': ['logistic', 'tanh', 'relu'],
    'solver': ['adam', 'sgd'],
    'alpha': [0, .2, .5, .7, 1],
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
    'learning_rate_init': [0.001, 0.01, 0.1, 0.2, 0.5],
    'max_iter': [5000]
}

ann = MLPClassifier()
grid_search = RandomizedSearchCV(estimator = ann, param_distributions=param_grid, cv=kfolds, n_iter=100,
                           scoring=score_measure, verbose=1, n_jobs=-1,  # n_jobs=-1 will utilize all available CPUs 
                           return_train_score=True)

_ = grid_search.fit(X_train, y_train)

bestacurracyann = grid_search.best_estimator_

print(grid_search.best_params_)
```

```python
%%time
y_pred = bestacurracyann.predict(X_test)

print(classification_report(y_test, y_pred))
```

## NN With GridSearchCV

```python
%%time

score_measure = "accuracy"
kfolds = 5

param_grid = {
    'hidden_layer_sizes': [ (30,), (50,), (70,), (90,)],
    'activation': ['tanh', 'relu'],
    'solver': ['adam'],
    'alpha': [.5, .7, 1],
    'learning_rate': ['adaptive', 'invscaling'],
    'learning_rate_init': [0.005, 0.01, 0.15],
    'max_iter': [5000]
}

ann = MLPClassifier()
grid_search = GridSearchCV(estimator = ann, param_grid=param_grid, cv=kfolds, 
                           scoring=score_measure, verbose=1, n_jobs=-1,  # n_jobs=-1 will utilize all available CPUs 
                           return_train_score=True)

_ = grid_search.fit(X_train, y_train)

bestacurracyann = grid_search.best_estimator_

print(grid_search.best_params_)
```

```python
%%time
y_pred = bestacurracyann.predict(X_test)

print(classification_report(y_test, y_pred))
```

# Decision tree with Default Parameters

```python
%%time
dtree=DecisionTreeClassifier()
_ = dtree.fit(X_train, y_train)
```

```python
%%time
y_pred = dtree.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

# Decision Tree (Hyperparameter) using Random Search

```python

%%time
# Create the random grid
param_grid = {
    'min_samples_split': np.arange(2,50),  
    'min_samples_leaf': np.arange(1,50),
    'min_impurity_decrease': np.arange(0.0001, 0.01, 0.0005),
    'max_leaf_nodes': np.arange(5, 200), 
    'max_depth': np.arange(1,50), 
    'criterion': ['entropy', 'gini'],
    
}

dtree=DecisionTreeClassifier()

best_random_search_model = RandomizedSearchCV(
        estimator=dtree, 
        scoring='accuracy', 
        param_distributions=param_grid, 
        n_iter = 200, 
        cv=5, 
        verbose=1, 
        n_jobs = -1
    )

_ = best_random_search_model.fit(X_train, y_train)
random_search_best_params = best_random_search_model.best_params_
print('Best parameters found: ', random_search_best_params)
print('Best score is {}'.format(best_random_search_model.best_score_))
```

```python
%%time
y_pred =best_random_search_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

# Decision Tree (Hyperparameter) using Grid Search

```python
%%time
plus_minus = 2
increment = 2

param_grid = { 'min_samples_split': [x for x in range(random_search_best_params['min_samples_split']-plus_minus, random_search_best_params['min_samples_split']+plus_minus,2) if x >= 2],       
              'min_samples_leaf': [x for x in range(random_search_best_params['min_samples_leaf']-plus_minus , random_search_best_params['min_samples_leaf']+plus_minus,2) if x > 0],
              'min_impurity_decrease': [x for x in np.arange(random_search_best_params['min_impurity_decrease']-0.001, random_search_best_params['min_impurity_decrease']+0.001,.0001).round(5) if x >= 0.000],
              'max_leaf_nodes':[x for x in range(random_search_best_params['max_leaf_nodes']-plus_minus , random_search_best_params['max_leaf_nodes']+plus_minus, 2) if x > 1],  
              'max_depth': [x for x in range(random_search_best_params['max_depth']-plus_minus , random_search_best_params['max_depth']+plus_minus, 2) if x > 1],
              'criterion': [random_search_best_params['criterion']]
              }
dtree=DecisionTreeClassifier()
best_grid_search_model = GridSearchCV(estimator=dtree, 
                                    scoring='accuracy', param_grid=param_grid, cv=5, verbose=1,  n_jobs = -1)
_ = best_grid_search_model.fit(X_train, y_train)
print('Best parameters found: ', best_grid_search_model.best_params_)
```

```python
%%time
y_pred = best_grid_search_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

# Interpretations
1.it seems that the ANN models perform significantly better than the Decision Tree models on the MNIST dataset. The default ANN model achieved an accuracy score of 0.97, which is quite high, and the ANN models that were fine-tuned using hyperparameter optimization techniques (random search and grid search) achieved even higher accuracy scores of 0.98, which is impressive.

2.The Decision Tree models, on the other hand, achieved lower accuracy scores. The default Decision Tree model achieved an accuracy score of 0.86, which is considerably lower than the default ANN model's score. However, the Decision Tree models that were fine-tuned using hyperparameter optimization techniques (random search and grid search) did improve in performance, achieving accuracy scores of 0.90 and 0.91, respectively.

3.we can see that the fit times and predict times vary across the different models and hyperparameter optimization techniques used.For the ANN models, the default model had a fit time of 688ms and a predict time of 6.93ms, while the random search and grid search models had significantly longer fit times of 7min 28s and 3min 30s, respectively. However, their predict times remained low at 10.7ms and 16.8ms, respectively.

4.For the Decision Tree models, the default model had a very short fit time of 28.7ms and a predict time of 9.08ms. The models that were fine-tuned using hyperparameter optimization techniques had longer fit times of 4.95s and 5.17s for the random search and grid search, respectively, but their predict times remained fast at 0ns.
So, it seems that the ANN models are a better choice for predicting handwritten numbers on the MNIST dataset, as they achieved higher accuracy scores than the Decision Tree models.


5.Overall, it seems that the Decision Tree models have faster fit and predict times than the ANN models, especially the default Decision Tree model, which had the shortest fit time of all models. However, the ANN models achieved higher accuracy scores, indicating better performance on the MNIST dataset. lets assume we focus on accuracy then computational efficiency we consider ANN grid model.
